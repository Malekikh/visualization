```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy 
from PIL import Image
import requests
from io import BytesIO
```

```python
x = np.array(['Luff', 'Zoro', 'Nami', 'USop', 'Sanji'])
y2 = np.array([170, 100, 90, 120, 50])
```

```python
response = requests.get('https://upload.wikimedia.org/wikipedia/en/c/cb/Monkey_D_Luffy.png') # Sends an HTTP GET request to the given URL to fetch the image.
image_file = BytesIO(response.content) # Wraps the binary content in a file-like object (BytesIO).
image = Image.open(image_file)  #Uses PIL.Image.open() to open the image
```

```python
fig, ax = plt.subplots(figsize = (7,3))
ax.plot(x,y2, color = 'red')

ax_image = fig.add_axes([0.1, 0.11, 0.15, 0.35]) #add a plot with given coordinates
ax_image.imshow(image) #retreive the image and locate on the plot
ax_image.axis('off') #remove borderline
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import scipy
import requests
import PIL
```

```python
tips = sns.load_dataset('tips') # load tip from seaborn library
print(tips)
```

```python
sns.set_style('whitegrid') # change style

tipgraph = sns.lineplot(data = tips,
             x = 'total_bill',
             y = 'tip',
            color = 'pink',
             linestyle = '--',
             linewidth = 3,
             marker = 'o',
             markerfacecolor = 'red')

tipgraph.set(title = 'Tips vs Total bill',
             xlabel = 'Total bill',
             ylabel = 'Tip amount')

tipgraph = sns.scatterplot(data = tips,
                           x = 'total_bill',
                           y = 'tip',
                           style = 'time',
                           hue = 'day',
                           palette = ['purple', 'red', 'yellow', 'blue'])
```

```python
fig = plt.subplot(figsize =(10,3))
```

```python
sns.pairplot(data = tips, hue = 'day') # explore relationship between variables 
```

```python
daysplot = sns.relplot(
data=tips,
x = 'total_bill',
y = 'tip',
hue = 'sex', # colored based on sex 
col = 'day', # colum based on day
kind = 'scatter', #type of data
col_wrap= 2 # number of columns 
)
```

```python
import plotly.graph_objects as go
x1 = np.array(['luf', 'zoro', 'nami', 'unsop', 'sanj'])
y1 = np.array([110, 180, 200, 90])
```

```python
graph = go.Figure()
graph.add_trace(go.Bar(x=x1, y=y1))

graph.update_layout(
title = 'pirate',
xaxis_title = 'pirate',
yaxis_title = 'score'
)

```

```python
graph = go.Figure()
graph.add_trace(go.Scatter(x=x1, y=y1,
                           mode = 'markers',
                           marker = dict(
                               size = 15,
                               color = 'hotpink',
                               opacity =1,
                               line = dict(width=5, color = 'purple')
                           )))

graph.update_layout(
title = 'pirate',
xaxis_title = 'pirate',
yaxis_title = 'score'
)
```

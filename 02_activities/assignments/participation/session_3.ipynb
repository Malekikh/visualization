```python
# load libraries
import pandas as pd
import numpy as np                                                                                                  
import matplotlib.pyplot as plt
import PIL
import requests
import scipy
```

```python
# random seed
np.random.seed(613)
# x is a range, and 2 y variables
x = np.arange(50)
y1 = np.random.randint(0,100,50)
y2 = np.random.randint(0,100,50)
```

```python
# plotting lines from x and y variables

fig, ax = plt.subplots(figsize = (5,3))

ax.plot(x,y1, label = "person-1")
ax.plot(x,y2, label = "person-1")
ax.legend(loc = 'lower right', # add arguments to legend
          frameon = True,
          fontsize = 10,
          ncol = 2,
          shadow = True) 
```

```python

x1 = np.arange(50)
y1 = np.random.randint(0,75,50)
x2 = np.array(["Luffy", 'Zoro',"nami", "Unsop", 'Sanji'])
y2 = np.array([110, 180, 240, 99, 220])

# plot 2 plots as subplots

fig, (ax1, ax2) = plt.subplots(ncols = 2, # relative position of subplots
                               nrows=1,
                               figsize =(9,3)
)
ax1.scatter(x1,y1, label = "Saved Lives", color = 'black')
ax2.bar(x2,y2, label = "Heros saved Lives", color = 'red')
ax1.legend(loc = 'lower left',
           bbox_to_anchor = (0,1))
ax2.legend(loc = 'lower left',
           bbox_to_anchor = (0,1))

ax1.grid()
```

```python
# add more subplots using subplot mosaic
fig, someaxes = plt.subplot_mosaic([['ax1', 'ax3'],
                                    ['ax2', 'ax3']], 
                                    figsize = (7,4),
                                    layout = "constrained") # ensure lables fit inside the frame

someaxes['ax1'].scatter(x1,y1)
someaxes['ax2'].bar(x2,y2)
someaxes['ax3'].plot(x1,y1)


someaxes['ax1'].set_xlabel('1st', fontsize = 18)
someaxes['ax2'].set_xlabel('2nd', fontsize = 18)
someaxes['ax3'].set_xlabel('3rd', fontsize = 18)
```

```python
x = np.array(["Luffy", 'Zoro',"nami", "Unsop", 'Sanji'])
y1 = np.array([110, 180, 240, 99, 220])
y2 = np.array([10, 180, 240, 99, 300])

fig, ax = plt.subplots(figsize = (7,3))
ax.bar(x,y1, color = 'black')
ax.plot(x,y2, color = 'red')
```

```python
y2_sd = np.std(y2)
fig, ax = plt.subplots(figsize = (7,3))

ax.plot(x,y2, color = 'red')

ax.errorbar(x,y2,
            yerr=y2_sd,
            fmt = 'none',
            ecolor = 'indigo',
            elinewidth = 4,
            capsize = 6,
            capthick = 6,
            errorevery=2
            )
```

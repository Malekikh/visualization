```python
import pandas as pd
import numpy as np                                                                                                  
import matplotlib.pyplot as plt
```

```python
np.random.seed(123)
x = np.arange(50)
y = np.random.randint(0,100,50)
```

```python
fig, ax = plt.subplots(figsize = (5,3))
ax.scatter(x,y)
```

```python
fig, ax = plt.subplots(figsize = (5,3))
ax.bar(x,y)
```

```python
fig, ax = plt.subplots(figsize = (5,3))
ax.plot(x,y)
```

```python
fig, ax = plt.subplots(figsize = (5,3))
ax.hist(y)
```

```python
font1 = {'family' : 'sans-serif', 'color' : 'blue', 'size' : 20}
font2 = {'family' : 'monospace', 'color' : 'green', 'size' : 14}
fig, ax = plt.subplots(figsize =(5,3))
ax.plot(x,y)
ax.set_title('total growth over time', fontdict = font1, loc ='left')
ax.set_ylabel('total growth', fontdict = font2)
ax.set_xlabel('years', fontdict = font2)


```

```python
fig, ax = plt.subplots(figsize =(5,3))
ax.scatter(x,y,
            marker = "*",
            color = "black")
```

```python
fig, ax = plt.subplots(figsize = (5,3))
ax.plot(x,y,
        marker = '*',
        color = "black",
        linestyle = "--",
        linewidth = 2,
        markersize = 12,
        markeredgecolor = 'orange',
        markerfacecolor = "red")
ax.grid(
    color = 'black',
    linewidth = 2,
    linestyle = '--')
```

```python
# libraries
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import pandas as pd
 
# Dataset:
a = pd.DataFrame({ 'group' : np.repeat('A',500), 'value': np.random.normal(10, 5, 500) })
b = pd.DataFrame({ 'group' : np.repeat('B',500), 'value': np.random.normal(13, 1.2, 500) })
c = pd.DataFrame({ 'group' : np.repeat('B',500), 'value': np.random.normal(18, 1.2, 500) })
d = pd.DataFrame({ 'group' : np.repeat('C',20), 'value': np.random.normal(25, 4, 20) })
e = pd.DataFrame({ 'group' : np.repeat('D',100), 'value': np.random.uniform(12, size=100) })
df = pd.concat([a,b,c,d,e])
 
# Usual boxplot
sns.boxplot(x='group', y='value', data=df, color= 'green')
plt.show()
```

```python
# import libraries
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
# open csv data
delays = pd.read_csv(r"C:\Users\khate\Downloads\TTC Streetcar Delay Data since 2025.csv")
```

```python
delays['min_delay'].describe()
```

```python
# filtering top 20 most delayed stations
top_stations = delays['station'].value_counts().nlargest(20).index
filtered_top_stations = delays[delays['station'].isin(top_stations)]
```

```python
# clean titles
delays = delays.rename(columns = {
    'Min Delay' : 'min_delay',
    'Station' : 'station',
    'Incident' : 'incident'
})
```

```python
# visualozation top 20 most delayed stations in 2025
plt.figure(figsize = (20,10))

sns.scatterplot(x = 'station', y = 'min_delay', hue='incident', data=filtered_top_stations)
plt.xticks(rotation=90)
plt.title('Top 20 Most Frequent Stations')
plt.show()
```
